```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Standardize the Variables
from sklearn.preprocessing import StandardScaler
```

```python
rs = 3717
```

```python
# ALL ITEMS INCLUDED
```

```python
scaler = StandardScaler()
scaler.fit(df.drop(['Patient\'s Age','BBS Sum','Assessment of Risk'],axis=1))
scaled_features = scaler.transform(df.drop(['Patient\'s Age','BBS Sum','Assessment of Risk'],axis=1))
df_feat = pd.DataFrame(scaled_features,columns=df.columns[1:-2])
df_feat.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Patient\'s Age','BBS Sum','Assessment of Risk'],axis=1),df['Assessment of Risk'],
                                                    test_size=0.30,random_state=rs)
```

```python
Xs_train, Xs_test, ys_train, ys_test = train_test_split(scaled_features,df['Assessment of Risk'],
                                                    test_size=0.30,random_state=rs)
```

```python
ero = []

for i in range(1,50):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    ero.append(np.mean(pred_i != y_test))
```

```python
ers = []

for i in range(1,50):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xs_train,ys_train)
    pred_i = knn.predict(Xs_test)
    ers.append(np.mean(pred_i != ys_test))
```

```python
cmpos = pd.concat([pd.DataFrame(ero,columns=['Original']),pd.DataFrame(ers,columns=['Scaled'])],axis=1)
cmpos.index += 1
cmpos.index.names = ['K']
cmpos.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,50),ero,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='green', markersize=7)
plt.plot(range(1,50),ers,color='red', linestyle='dashed', marker='o',
         markerfacecolor='yellow', markersize=7)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# USING ALL ITEMS ORIGINAL(UNSCALED) is BEST, and K = 3 is BEST
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
kns = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
kns.fit(Xs_train,ys_train)
pres = kns.predict(Xs_test)
print('Original (Unscaled)')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
print('\n')
print('Scaled')
print(confusion_matrix(ys_test,pres))
print('\n')
print(classification_report(ys_test,pres))
```
